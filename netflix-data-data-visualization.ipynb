```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/netflix-data-cleaning-analysis-and-visualization/netflix1.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates
```

```python
df.columns
```

```python
type_counts = df.groupby('type').size()

type_counts.plot.pie(autopct='%.1f%%')
plt.title('Count of Type')
plt.ylabel('')
plt.show()

```

```python
country_counts = df.groupby('country')['title'].nunique().nlargest(10)

country_counts.plot.bar()
plt.xlabel('Country')
plt.ylabel('Count of Titles')
plt.title('Top 10 Countries by Count of Titles')
plt.show()
```

```python
rating_counts = df.groupby('rating')['title'].nunique().nlargest(14)

rating_counts.plot.bar()
plt.xlabel('rating')
plt.ylabel('Count of Titles')
plt.title('Count of Titles per Rating')
plt.show()
```

```python
listed_in_counts = df.groupby('listed_in')['title'].nunique().nlargest(10)
sns.set(rc={'figure.figsize':(10, 5)})
sns.barplot(x=listed_in_counts.index, y=listed_in_counts.values)
plt.xlabel('listed_in')
plt.ylabel('Count of Titles')
plt.title('Top 10 listed_in by Count of Titles')
plt.show()
```

```python
release_year_counts = df.groupby('release_year')['title'].nunique().nlargest(10)
sns.set(rc={'figure.figsize':(10, 5)})
sns.lineplot(x=release_year_counts.index, y=release_year_counts.values)
plt.xlabel('release_year')
plt.ylabel('Count of Titles')
plt.title('Top 10 Release Year by Count of Titles')
plt.show()
```
